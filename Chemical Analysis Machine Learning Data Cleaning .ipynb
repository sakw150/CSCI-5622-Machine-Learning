```r
CAdata=read.csv(file="~/Desktop/Machine Learning Wine Dataset/ChemicalAnalysis.csv", sep=",", header=TRUE)
head(CAdata) # Original Dataset Variables List 
summary(CAdata) # Original Dataset Summary 
# When initially viewing the data, it seems like there is no incomplete datasets. All of them are full sets.
# Therefore, we are going to analyze each column to determine any outliers. The outliers will be removed. 
names(CAdata)
nrow(CAdata)
```

```r
# Alcohol  (No Outliers Present)
summary(CAdata$Alcohol)
length(CAdata$Alcohol)
par(mfrow = c(1, 3))
hist(CAdata$Alcohol, main = "Histogram")
boxplot(CAdata$Alcohol, main = "Boxplot")
qqnorm(CAdata$Alcohol, main = "Normal Q-Q plot")
```

```r
# Malic Acid (Outliers Present and Some Removed. Some will stay as noise)
summary(CAdata$Malic_Acid)
par(mfrow = c(1, 3))
hist(CAdata$Malic_Acid, main = "Histogram")
boxplot(CAdata$Malic_Acid, main = "Boxplot")
qqnorm(CAdata$Malic_Acid, main = "Normal Q-Q plot")

IQR(CAdata$Malic_Acid) #Tmin=Q1-(c*IQR), Tmax=Q3+(c*IQR)
Tmin=1.603-(1.5*1.48)
Tmax=3.083+(1.5*1.48)
CAdata2=CAdata[which(CAdata$Malic_Acid > Tmin & CAdata$Malic_Acid < Tmax),]
par(mfrow = c(1, 3))
hist(CAdata2$Malic_Acid, main = "Histogram")
boxplot(CAdata2$Malic_Acid, main = "Boxplot")
qqnorm(CAdata2$Malic_Acid, main = "Normal Q-Q plot")
```

```r
# Ash  (Outliers Present and some removed. Some will stay as noise)
summary(CAdata2$Ash)
par(mfrow = c(1, 3))
hist(CAdata2$Ash, main = "Histogram")
boxplot(CAdata2$Ash, main = "Boxplot")
qqnorm(CAdata2$Ash, main = "Normal Q-Q plot")

IQR(CAdata2$Ash) #Tmin=Q1-(c*IQR), Tmax=Q3+(c*IQR)
Tmin=2.210-(1.5*0.345)
Tmax=2.555+(1.5*0.345)
CAdata3=CAdata2[which(CAdata2$Ash > Tmin & CAdata2$Ash < Tmax),]
par(mfrow = c(1, 3))
hist(CAdata3$Ash, main = "Histogram")
boxplot(CAdata3$Ash, main = "Boxplot")
qqnorm(CAdata3$Ash, main = "Normal Q-Q plot")
```

```r
# Ash_Alcanity (Outliers Present and Removed)
summary(CAdata3$Ash_Alcanity)
par(mfrow = c(1, 3))
hist(CAdata3$Ash_Alcanity, main = "Histogram")
boxplot(CAdata3$Ash_Alcanity, main = "Boxplot")
qqnorm(CAdata3$Ash_Alcanity, main = "Normal Q-Q plot")

IQR(CAdata3$Ash_Alcanity) #Tmin=Q1-(c*IQR), Tmax=Q3+(c*IQR)
Tmin=17.18-(1.5*4.325)
Tmax=21.50+(1.5*4.325)
CAdata4=CAdata3[which(CAdata3$Ash_Alcanity > Tmin & CAdata3$Ash_Alcanity < Tmax),]
par(mfrow = c(1, 3))
hist(CAdata4$Ash_Alcanity, main = "Histogram")
boxplot(CAdata4$Ash_Alcanity, main = "Boxplot")
qqnorm(CAdata4$Ash_Alcanity, main = "Normal Q-Q plot")
```

```r
# Magnesium (Outliers Present and Removed)
summary(CAdata4$Magnesium)
par(mfrow = c(1, 3))
hist(CAdata4$Magnesium, main = "Histogram")
boxplot(CAdata4$Magnesium, main = "Boxplot")
qqnorm(CAdata4$Magnesium, main = "Normal Q-Q plot")

IQR(CAdata4$Magnesium) #Tmin=Q1-(c*IQR), Tmax=Q3+(c*IQR)
Tmin=88-(1.5*19)
Tmax=107+(1.5*19)
CAdata5=CAdata4[which(CAdata4$Magnesium > Tmin & CAdata4$Magnesium < Tmax),]
par(mfrow = c(1, 3))
hist(CAdata5$Magnesium, main = "Histogram")
boxplot(CAdata5$Magnesium, main = "Boxplot")
qqnorm(CAdata5$Magnesium, main = "Normal Q-Q plot")
```

```r
# Proanthocyanins (Some Outliers Removed and Some Sound Present)
summary(CAdata5$Proanthocyanins)
par(mfrow = c(1, 3))
hist(CAdata5$Proanthocyanins, main = "Histogram")
boxplot(CAdata5$Proanthocyanins, main = "Boxplot")
qqnorm(CAdata5$Proanthocyanins, main = "Normal Q-Q plot")

IQR(CAdata5$Proanthocyanins) #Tmin=Q1-(c*IQR), Tmax=Q3+(c*IQR)
Tmin=1.25-(1.5*0.675)
Tmax=1.925+(1.5*0.675)
CAdata6=CAdata5[which(CAdata5$Proanthocyanins > Tmin & CAdata5$Proanthocyanins < Tmax),]
par(mfrow = c(1, 3))
hist(CAdata6$Proanthocyanins, main = "Histogram")
boxplot(CAdata6$Proanthocyanins, main = "Boxplot")
qqnorm(CAdata6$Proanthocyanins, main = "Normal Q-Q plot")
```
