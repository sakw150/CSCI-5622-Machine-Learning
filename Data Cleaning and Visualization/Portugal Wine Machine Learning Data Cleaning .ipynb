```r
PWdata=read.csv(file="~/Desktop/Machine Learning Module 1/PortugalWine.csv", sep=",", header=TRUE)
head(PWdata) # Original Dataset Variables List 
summary(PWdata) # Original Dataset Summary 
nrow(PWdata)

# Adjusted Dataset (Removed Free.sulfur dioxide cause total.sulfur.dioxide is a combination of free.sulfur dioxide)
PWdata2=subset(PWdata, select = -c(free.sulfur.dioxide))
head(PWdata2)
names(PWdata)
```

```r
# Fixed Acidity
summary(PWdata2$fixed.acidity)
par(mfrow = c(1, 3))
hist(PWdata2$fixed.acidity, main = "Histogram")
boxplot(PWdata2$fixed.acidity, main = "Boxplot")
qqnorm(PWdata2$fixed.acidity, main = "Normal Q-Q plot")

IQR(PWdata2$fixed.acidity) #Tmin=Q1-(c*IQR), Tmax=Q3+(c*IQR)
Tmin=7.1-(1.5*2)
Tmax=9.1+(1.5*2)

PWdata3=PWdata2[PWdata2$fixed.acidity >Tmin & PWdata2$fixed.acidity <Tmax, ]
par(mfrow = c(1, 3))
hist(PWdata3$fixed.acidity, main = "Histogram")
boxplot(PWdata3$fixed.acidity, main = "Boxplot")
qqnorm(PWdata3$fixed.acidity, main = "Normal Q-Q plot")
```
